# 2. Run Regression Model

Population Regressio Model: 

$log(salary) = \beta_0 + \beta_1 log(sales) +u$

```python
ceosal1 = woo.dataWoo('ceosal1') # import data ceosal1 from woo
ceosal1.columns
```

## 2.1. Log transformation: Take the log of variables

```python
ceosal1['log_salary'] = np.log(ceosal1['salary']) # create log(salary)
ceosal1['log_sales'] = np.log(ceosal1['sales']) # create log(sales)
```

## 2.2. Descriptive Statistics

```python
ceosal1[['log_salary', 'log_sales']].describe() # get descritive statistics
```

```python
ceosal1[['log_salary', 'log_sales']].cov() # check correlation between y and x variables
```

## 2.3. Run a regression model

```python
import statsmodels.formula.api as smf # import package statsmodels for regression model
import statsmodels.api as sm # create other analysis in statsmodels
```

```python
reg = smf.ols(formula='log_salary ~ log_sales', data=ceosal1)
results = reg.fit() # regression results
print(results.summary())
```

```python
table0=sm.stats.anova_lm(results, type=2) # create a table with SSE and SSR
print(table0)
```

### Calculate SST with the above information

```python
print("SSE: ", table0['sum_sq'][0])
print("SSR: ", table0['sum_sq'][1])
print("SST: ", table0['sum_sq'][0] + table0['sum_sq'][1])
```

# 3. Multiple Linear Regression Model 

```python
gpa1 = woo.data("gpa1") # load data 
gpa1
```

```python
ols_gpa = smf.ols('colGPA ~ hsGPA + ACT', data = gpa1)
result5 = ols_gpa.fit()
print(result5.summary())
```

## 3.1. Check Multicollinearity

```python
# create hsGPA2 
from numpy import random
len(gpa1)
```

```python
ols_gpa = smf.ols('colGPA ~ hsGPA + hsGPA2 + ACT', data = gpa1)
result6 = ols_gpa.fit()
print(result6.summary())
```
